```python
import pandas as pd
from sqlalchemy import create_engine, inspect
import json
pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv("Downloads/executions.csv")
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month 
```

```python
df['County & State'] = df['County'].astype(str) +", "+ df["State"]
```

```python
df.info()
```

```python
df.head()
```

```python
dfc = pd.read_excel('Downloads/us-county-boundaries.xlsx')
```

```python
dfc['County & State'] = dfc['County'].astype(str) +", "+ dfc["State"]
```

```python
dfc.head()
```

```python
df2 = pd.merge(df, dfc, on="County & State")
```

```python
df2.columns
```

```python
df3 = df2.loc[:,['Date', 'Month', 'Year', 'Name', 'Age', 'Sex', 'Race', 'Victim Count',  'Method', 'County_x', 'State_x', 'Coordinates']]
```

```python
df3 = df3.rename(columns={"County_x": "County", "State_x": "State"})
```

```python
df3.info()
```

```python
path = "sqlite:///data/executions_with_coordinates.sqlite"
engine = create_engine(path)
```

```python
df3.to_sql("executions_with_coordinates", con=engine, method="multi", if_exists="replace", index=False)
```

```python
# INSPECT
inspector_gadget = inspect(engine)

# get all tables
table_names = inspector_gadget.get_table_names()
for table in table_names:
    print(table)
    
    # get columns for table
    for column in inspector_gadget.get_columns(table):
        print(column)
        
    print()
```
